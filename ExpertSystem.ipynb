Salah satu tahapan inti dalam skincare adalah pemakaian serum, oleh karenanya pemilihan serum yang sesuai kebutuhan harus diprioritaskan agar tidak menagalami iritasi akibat pemakaian serum dan mendapatkan hasil yang di inginkan.


Beberapa jenis kandungan serum wajah: 

*   Hyaluronic Acid
*   Niacinamide
*   AHA & BHA
*   Marine Collagen
*   Lactic Acid
*   Snail Truecica
*   Salicylic Acid
*   Retinol
*   Vit c dan e
*   Honey (propolis)
*   tea (tea tree oil/ matcha )

```python
from experta import * 
import sys
import ast


class MedicalExpert(KnowledgeEngine):
    username = "", 


    @DefFacts()
    def needed_data(self):
        """ 
        This is a method which is called everytime engine.reset() is called.
        It acts like a constructor to this class.
        """        
        yield Fact(findDisease = 'true')
        print("Halo!Saya adalah Mr.Expert.\n\n Kamu dapat menemukan jenis serum yang sesuai untukmu!\n Aku akan mengajukan beberapa pertanyaan.\n\n")
        

    @Rule(Fact(findDisease = 'true'),NOT(Fact(name=W())),salience = 1000)
    def ask_name(self):
        self.username = input("Siapa Namamu?\n")
        self.declare(Fact(name=self.username))

    @Rule(Fact(findDisease='true'), NOT (Fact(jerawat = W())),salience = 995)
    def hasChestPain(self):
        self.jerawat_ = input("\nApakah Anda memiliki masalah jerawat?\nKetikkan Yes/No\n")
        self.jerawat_ = self.jerawat_.lower()
        self.declare(Fact(jerawat = self.jerawat_.strip().lower()))

  
    
    @Rule(Fact(findDisease='true'), NOT (Fact(bopeng = W())),salience = 985)
    def hasCough(self):
        self.bopeng = input("\nApakah kamu memiliki bopeng di wajah?\nKetikkan Yes/No\n")
        self.bopeng = self.bopeng.lower()
        self.declare(Fact(cough = self.bopeng.strip().lower()))

    



    @Rule(Fact(findDisease='true'), NOT (Fact(bekasj = W())),salience = 975)
    def hasFainting(self):
        self.bekasj = input("\nApakah kamu ingin menghilangkan bekas jerawat?\nKetikkan Yes/No\n")
        self.bekasj = self.bekasj.lower()
        self.declare(Fact(bekasj = self.bekasj.strip().lower()))


  

    @Rule(Fact(findDisease='true'), NOT (Fact(keriput = W())),salience = 970)
    def hasFatigue(self):
        self.keriput = input("\nApakah kamu memiliki kerutan di wajah?\nKetikkan Yes/No\n")
        self.keriput = self.keriput.lower()
        self.declare(Fact(keriput = self.keriput.strip().lower()))

    @Rule(Fact(findDisease='true'), NOT (Fact(kusam = W())),salience = 965)
    def hasHeadache(self):
        self.kusam = input("\nApakah kamu memiliki kulit kusam?\nKetikkan Yes/No\n")
        self.kusam = self.kusam.lower()
        self.declare(Fact(kusam = self.kusam.strip().lower()))
    

    @Rule(Fact(findDisease='true'), NOT (Fact(kulitkering = W())),salience = 955)
    def hasbackPain(self):
        self.kulitkering = input("\nApakah kamu memiliki kulit kering?\nKetikkan Yes/No\n")
        self.kulitkering = self.kulitkering.lower()
        self.declare(Fact(kulitkering = self.kulitkering.strip().lower()))
    
    @Rule(Fact(findDisease='true'), NOT (Fact(kulitminyak = W())),salience = 950)
    def hasSunkenEyes(self):
        self.kulitminyak = input("\nApakah kamu memiliki kulit berminyak?\nKetikkan Yes/No\n")
        self.kulitminyak = self.kulitminyak.lower()
        self.declare(Fact(kulitminyak = self.kulitminyak.strip().lower()))

    @Rule(Fact(findDisease='true'), NOT (Fact(sensitif = W())),salience = 945)
    def hasfever(self):
        self.sensitif = input("\nApakah Kamu memiliki kulit sensitif?\nPlease type Yes/No\n")
        self.sensitif=self.sensitif.lower()
        self.declare(Fact(sensitif = self.sensitif.strip().lower()))

    @Rule(Fact(findDisease='true'), NOT (Fact(awal = W())),salience = 940)
    def hassorethroat(self):
        self.awal = input("\nApakah kamu belum pernah menggunakan skincare sebelumnya?\nKetikkan Yes/No\n")
        self.awal = self.awal.lower()
        self.declare(Fact(awal = self.awal.strip().lower()))


    @Rule(Fact(findDisease='true'), NOT (Fact(muda = W())),salience = 935)
    def hasrestlessness(self):
        self.muda = input("\nApakah kamu ingin terlihat tampak lebih muda?\nKetikkan Yes/No\n")
        self.muda = self.muda.lower()
        self.declare(Fact(muda = self.muda.strip().lower()))


    @Rule(Fact(findDisease='true'),Fact(jerawat = 'yes'), Fact(bopeng = 'no'), Fact(bekasj = 'no'),Fact(keriput = 'no'),
    Fact(kusam = 'no'),Fact(kulitkering = 'no'),Fact(kulitminyak = 'yes'),Fact(sensitif = 'no'),Fact(awal='yes'),
    Fact(muda = 'no'))
    def disease_0(self):
        self.declare(Fact(disease = 'Salicylic Acid'))

    @Rule(Fact(findDisease='true'),Fact(jerawat = 'no'), Fact(bopeng = 'yes'), Fact(bekasj = 'no'),Fact(keriput = 'no'),
    Fact(kusam = 'no'),Fact(kulitkering = 'yes'),Fact(kulitminyak = 'no'),Fact(sensitif = 'no'),Fact(awal='yes'),
    Fact(muda = 'no'))
    def disease_1(self):
        self.declare(Fact(disease = 'Snail Truecica'))

    @Rule(Fact(findDisease='true'),Fact(jerawat = 'no'), Fact(bopeng = 'no'), Fact(bekasj = 'yes'),Fact(keriput = 'no'),
    Fact(kusam = 'yes'),Fact(kulitkering = 'yes'),Fact(kulitminyak = 'no'),Fact(sensitif = 'no'),Fact(awal='yes'),
    Fact(muda = 'no'))
    def disease_2(self):
        self.declare(Fact(disease = 'Niacinamide'))

    @Rule(Fact(findDisease='true'),Fact(jerawat = 'no'), Fact(bopeng = 'no'), Fact(bekasj = 'no'),Fact(keriput = 'yes'),
    Fact(kusam = 'yes'),Fact(kulitkering = 'yes'),Fact(kulitminyak = 'yes'),Fact(sensitif = 'yes'),Fact(awal='yes'),
    Fact(muda = 'yes'))
    def disease_3(self):
        self.declare(Fact(disease = 'Hyaluronic Acid'))


    @Rule(Fact(findDisease='true'),Fact(jerawat = 'no'), Fact(bopeng = 'yes'), Fact(bekasj = 'yes'),Fact(keriput = 'yes'),
    Fact(kusam = 'no'),Fact(kulitkering = 'yes'),Fact(kulitminyak = 'no'),Fact(sensitif = 'no'),Fact(awal='no'),
    Fact(muda = 'yes'))
    def disease_4(self):
        self.declare(Fact(disease = 'Marine Collagen'))


    @Rule(Fact(findDisease='true'),Fact(jerawat = 'no'), Fact(bopeng = 'no'), Fact(bekasj = 'yes'),Fact(keriput = 'no'),
    Fact(kusam = 'yes'),Fact(kulitkering = 'no'),Fact(kulitminyak = 'no'),Fact(sensitif = 'no'),Fact(awal='no'),
    Fact(muda = 'no'))
    def disease_5(self):
        self.declare(Fact(disease = 'Vit C dan E'))

    @Rule(Fact(findDisease='true'),Fact(jerawat = 'yes'), Fact(bopeng = 'no'), Fact(bekasj = 'yes'),Fact(keriput = 'yes'),
    Fact(kusam = 'yes'),Fact(kulitkering = 'yes'),Fact(kulitminyak = 'yes'),Fact(sensitif = 'no'),Fact(awal='no'),
    Fact(muda = 'yes'))
    def disease_6(self):
        self.declare(Fact(disease = 'Retinol'))

    @Rule(Fact(findDisease='true'),Fact(jerawat = 'no'), Fact(bopeng = 'no'), Fact(bekasj = 'yes'),Fact(keriput = 'no'),
    Fact(kusam = 'yes'),Fact(kulitkering = 'yes'),Fact(kulitminyak = 'yes'),Fact(sensitif = 'yes'),Fact(awal='yes'),
    Fact(muda = 'no'))
    def disease_7(self):
        self.declare(Fact(disease = 'Honey (propolis)'))

    @Rule(Fact(findDisease='true'),Fact(jerawat = 'yes'), Fact(bopeng = 'no'), Fact(bekasj = 'yes'),Fact(keriput = 'no'),
    Fact(kusam = 'no'),Fact(kulitkering = 'no'),Fact(kulitminyak = 'yes'),Fact(sensitif = 'no'),Fact(awal='yes'),
    Fact(muda = 'no'))
    def disease_8(self):
        self.declare(Fact(disease = 'tea (tea tree oil/ matcha'))
    

    @Rule(Fact(findDisease='true'),NOT (Fact(disease = W())),salience = -1)
    def unmatched(self):
        self.declare(Fact(disease = 'unknown'))

    @Rule(Fact(findDisease = 'true'),Fact(disease = MATCH.disease),salience = 1)
    def getDisease(self, disease):
        
        if(disease == 'unknown'):
            mapDisease = []
            mapDisease.append('kulitkering')
            mapDisease.append('jerawat_')
            mapDisease.append('bopeng')
            mapDisease.append('bekasj')
            mapDisease.append('keriput')
            mapDisease.append('sensitif')
            mapDisease.append('kusam')
            mapDisease.append('awal')
            mapDisease.append('muda')
            mapDisease.append('kulitminyak') 
            print('\n\nWe checked the following symptoms',mapDisease)
            mapDisease_val=[self.kulitkering,self.jerawat_,self.bopeng,self.bekasj,self.keriput
            ,self.sensitif,self.kusam,self.awal,self.muda,self.kulitminyak]
            print('\n\nSymptoms in patients are :', mapDisease_val)
            
            file = open('disease_symptoms.txt','r')
            contents = file.read()
            dictionary = ast.literal_eval(contents)
            file.close()
            
            yes_symptoms = []
            for i in range(0,len(mapDisease_val)):
                if mapDisease_val[i] == 'yes':
                    yes_symptoms.append(mapDisease[i])
            
            max_val = 0
            print('\n\nYes symptoms noticed are : ', yes_symptoms)
            for key in dictionary.keys():
                val = dictionary[key].split(",")
                count = 0
                print(key,":",val)
                for x in val:
                    if x in yes_symptoms:
                        count+=1
                
                if count > max_val:
                    max_val = count
                    pred_dis = key
            
            if max_val == 0:
                print("\n# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #")
                print("Tidak dapat menemukan serum yang cocok, cobalah skincare dasar terlebih dahulu")
                print("\n# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #")
            else:
                print("\n# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #")
                print("\n\nKami tidak dapat memastikan serum secara pasti.\nTetapi kami merekomendasikan serum yang cocok adalah ",pred_dis)
                print("\n\n")
                print("\n# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #")
                
                
        else:
            print("\n# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #")
            print('Serum yang cocok buatmu adalah:',disease)
            print('\n\n')
            print("\n# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #")
            
if __name__ == "__main__":
    engine = MedicalExpert()
    engine.reset()
    engine.run()
    print('Printing engine facts after 1 run',engine.facts)
   
```
